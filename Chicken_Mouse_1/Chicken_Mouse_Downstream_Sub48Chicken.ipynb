# Importing Samap Data

```python
with open('samap_object_sub48.pkl', 'rb') as f:
    data = pickle.load(f)
print(data)
sm = data
samap = sm.samap
```

# Analysis By Treatment

```python
keys = {'ch':'orig.ident','ms':'active_ident'}
D,MappingTable = get_mapping_scores(sm,keys,n_top = 0)
D.head()
```

```python
sankey_plot(MappingTable, align_thr=0.05, species_order = ['ch','ms'])
```

```python
chord_plot(MappingTable, align_thr=0.05)
```

```python
gpfIdent = GenePairFinder(sm,keys=keys)
gene_pairsident = gpfIdent.find_all(align_thr=0.2)
gene_pairsident.head()
```

```python
csv = gene_pairsident.to_csv()
```

```python
with open("treatmentgenepairs.csv", "w") as file:
    file.write(csv)
    file.close()
```
